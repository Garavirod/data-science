```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv('../simple/datasets/auto-mpg.csv')
data_frame.head()
```

```python
data_frame.shape
```

```python
%matplotlib inline
data_frame['mpg'] = data_frame['mpg'].dropna()
data_frame['horsepower'] = data_frame['horsepower'].dropna()
x_plot, y_plot = data_frame['horsepower'], data_frame['mpg']
plt.scatter(x=x_plot, y=y_plot, color="navy")
plt.ylabel('Consume by galeon')
plt.xlabel('PH')
plt.title('Galeon intake VS PH')
```

```python
# linear regression model

X = data_frame['horsepower'].fillna(data_frame['horsepower'].mean())
X
```

```python

Y = data_frame['mpg'].fillna(data_frame['mpg'].mean())
Y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=0)

X_RESHAPED = np.array(X).reshape(-1,1)
x_train_reshaped = np.array(x_train).reshape(-1,1)
x_test_reshaped = np.array(x_test).reshape(-1,1)

lm = LinearRegression()
lm.fit(x_train_reshaped,y_train)

y_prediction_testing = lm.predict(x_test_reshaped)
```

```python
plt.scatter(x=X, y=Y, color='navy')
plt.plot(x_test, y_prediction_testing, color='red')
```

```python
lm.score(X_RESHAPED, Y)
```
