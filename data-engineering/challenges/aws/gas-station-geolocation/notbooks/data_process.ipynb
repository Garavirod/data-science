Mapbox API

- [Mapbox geocoding](https://docs.mapbox.com/help/getting-started/geocoding/)

```python
from dotenv import load_dotenv
import requests
load_dotenv()

def get_coords_from_address(address:str):
    address_to_send = f'{address}, Mexico City, Mexico'
    api_key = os.getenv('MAPBOX_API_KEY')
    URL = f'https://api.mapbox.com/geocoding/v5/mapbox.places/{address_to_send}.json?types=address&access_token={api_key}'
    try:
        response = requests.get(URL).json()
        return response['features'][0]['center']
    except:
        print("Not Found")
        return None
    
def calculate_distance(point1, point2, unit):
    if unit == 'km':
        distance = haversine(point1,point2, Unit.KILOMETERS)
    elif unit == 'm':
        distance = haversine(point1,point2, Unit.METERS)
    else:
        distance = haversine(point1,point2, Unit.MILES)

    return round(distance,2)
```

```python
coords = get_coords_from_address(address='Palacio de Bellas Artes')
coords
```

```python
folium_map_instance = folium.Map([coords[1],coords[0]], zoom_start=15)
# Add circle to map
folium.Circle(
    location=[coords[1],coords[0]],
    color='green',
    fill = 'green',
    radius=1000
).add_to(folium_map_instance)


# Add marker
folium.Marker(
    location=[coords[1],coords[0]],
    icon=folium.Icon(color='black', icon_color='white', icon='home', prefix='glyphicon'),
    popup='<h5>Im here</h5>'
).add_to(folium_map_instance)

folium_map_instance
```
