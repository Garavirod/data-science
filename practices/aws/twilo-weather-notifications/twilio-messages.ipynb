### Building the URL

```python
query = 'Mexico'
api_key = API_KEY_WAPI
api_weather_url = f'http://api.weatherapi.com/v1/forecast.json?key={api_key}&q={query}&days=1&aqi=no&alerts=no'
```

### Getting api response

```python
api_response = requests.get(url=api_weather_url).json()
api_response
```

```python
# Getting key from json
dict_keys = api_response.keys()
dict_keys
```

```python
# Getting forecaste data
forecast_data = api_response['forecast']['forecastday'][0].keys()
print(forecast_data)

# date
date = api_response['forecast']['forecastday'][0]['hour'][0]['time'].split()[0]
print(date)
# hour
hour = int(api_response['forecast']['forecastday'][0]['hour'][0]['time'].split()[1].split(':')[0])
print(hour)
# condition
condition = api_response['forecast']['forecastday'][0]['hour'][0]['condition']['text']
print(condition)
# temperature
temperature = api_response['forecast']['forecastday'][0]['hour'][1]['temp_c']
print(temperature)
# will it rain?
will_rain = api_response['forecast']['forecastday'][0]['hour'][0]['will_it_rain']
print(will_rain)
# chance of rain
chance_rain = api_response['forecast']['forecastday'][0]['hour'][2]['chance_of_rain']
print(chance_rain)
```
